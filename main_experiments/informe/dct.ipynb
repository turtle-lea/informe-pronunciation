```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import dct
```

```python
two_pi = 2 * np.pi
cos = np.cos
```

```python
def cosine(f, t):
    return cos(two_pi * f * t)
```

## Match 1Hz

```python
f = 1
```

```python
x0 = 0
x1 = two_pi / 2
xs = np.array([x0, x1])
ys = cos(xs)
```

```python
coeffs = dct(ys, norm="ortho")
a_1 = coeffs[1]
```

```python
computed_cosine = lambda t: a_1 * cos(two_pi*1*t)
```

```python
new_xs = np.linspace(0, two_pi * 2, 100)/two_pi
new_ys = computed_cosine(new_xs)
```

```python
base_xs = np.linspace(0, two_pi * 2, 100)/two_pi
base_ys = cosine(1, base_xs)
```

```python
sns.set(style="whitegrid")
```

```python
plt.plot(xs/two_pi, ys, marker="o", linestyle="None", color="black")
plt.plot(base_xs, base_ys, color="black", label="Baseline", linestyle = "--")
plt.plot(new_xs, new_ys, color="blue", label="Computed")
plt.legend()
plt.show()
```

## Match 2Hz

```python
def cosine(f, t):
    return cos(two_pi * f * t)
```

```python
f = 2
```

```python
x0 = 0
x1 = two_pi/3
x2 = 2 * two_pi/3 
xs = np.array([x0, x1, x2])
ys = cos(xs)
```

```python
coeffs = dct(ys, norm="ortho")
a_1 = coeffs[1]
a_2 = coeffs[2]
```

```python
computed_cosine = lambda t: a_1 * cos(two_pi*1*t) + a_2 * cos(two_pi*2*t)
```

```python
new_xs = np.linspace(0, two_pi * 2, 100)/two_pi
new_ys = computed_cosine(new_xs)
```

```python
base_xs = np.linspace(0, two_pi * 2, 100)/two_pi
base_ys = cosine(2, base_xs)
```

```python
sns.set(style="whitegrid")
plt.plot(xs/two_pi, ys, marker="o", linestyle="None", color="black")
plt.plot(base_xs, base_ys, color="black", label="Baseline", linestyle = "--")
plt.plot(new_xs, new_ys, color="blue", label="Computed")
plt.legend()
plt.show()
```

## Match 3Hz

```python
def cosine(f, t):
    return cos(two_pi * f * t)
```

```python
f = 3
```

```python
x0 = 0
x1 = two_pi/4
x2 = 2 * two_pi/4
x3 = 3 * two_pi/4
xs = np.array([x0, x1, x2, x3])
ys = cos(xs)
```

```python
coeffs = dct(ys, norm="ortho")
a_1 = coeffs[1]
a_2 = coeffs[2]
a_3 = coeffs[3]
```

```python
computed_cosine = lambda t: a_1 * cos(two_pi*1*t) + a_2 * cos(two_pi*2*t) + a_3 * cos(two_pi*3*t)
```

```python
new_xs = np.linspace(0, 1, 100)
new_ys = computed_cosine(new_xs)
```

```python
base_xs = np.linspace(0, 1, 100)
base_ys = cosine(3, base_xs)
```

```python
sns.set(style="whitegrid")
plt.plot(xs/two_pi, ys, marker="o", linestyle="None", color="black")
plt.plot(base_xs, base_ys, color="black", label="Baseline", linestyle = "--")
plt.plot(new_xs, new_ys, color="blue", label="Computed")
plt.legend()
plt.show()
```

## Match 1Hz + 2Hz

```python
def cosine_with_2(f1, f2, t):
    return (cos(two_pi * f1 * t) + cos(two_pi * f2 * t))
```

```python
def cosine_with_3(f1, f2, f3, t):
    return (cos(two_pi * f1 * t) + cos(two_pi * f2 * t)) + cos(two_pi * f3 * t) 
```

```python
def cosine_with_4(f1, f2, f3, f4, t):
    return (cos(two_pi * f1 * t) + cos(two_pi * f2 * t)) + cos(two_pi * f3 * t) + cos(two_pi * f4 * t)
```

```python
def cosine_with_5(f1, f2, f3, f4, f5, t):
    return (cos(two_pi * f1 * t) + cos(two_pi * f2 * t)) + cos(two_pi * f3 * t) + cos(two_pi * f4 * t) + cos(two_pi * f5 * t)
```

```python
def cosine_with_6(f1, f2, f3, f4, f5, f6, t):
    return (cos(two_pi * f1 * t) + cos(two_pi * f2 * t)) + cos(two_pi * f3 * t) + cos(two_pi * f4 * t) + cos(two_pi * f5 * t) + cos(two_pi * f6 * t)
```
